```python
import numpy as np
import pandas as pd
import h5py
from keras.models import load_model
from keras.optimizers import SGD
```

```python
# Load data
with h5py.File("data/data.hdf5", "r") as f:
    X = f["X"][:]
    X.shape = (377414, 1, 200, 12)   # Reshaped to fit model
    artist = f["artist"][:]
    song = f["song"][:]
    Y = f["Y"][:]
```

```python
# Load model
model = load_model('model.h5')
```

```python
# Get recommendations
def mse(x,y):
    return sum(sum((x-y)**2))
idx = 10000   # Song number
pred_song = X[idx]   
print(artist[idx])
print(song[idx])
pred_song.shape = (1,1,200,12)
prediction = model.predict(pred_song)
mses = [mse(prediction,i) for i in Y]
data = pd.DataFrame({"songs":song,"mses":mses,"artist":artist})
data.head()
data.sort(columns=["mses"])
```
